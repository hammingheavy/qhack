```python
import json
import pennylane as qml
import pennylane.numpy as np
import qiskit
```

```python
dev = qml.device("default.qubit", wires=5)
```

```python
@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=0)
    qml.QFT(wires=[2, 1, 0])
    qml.QFT(wires=[2, 1, 0])
    qml.QFT(wires=[4, 3, 2])
    qml.QFT(wires=[4, 3, 2]) 
    return qml.probs(wires=[2, 1, 0]), qml.state()

circuit()
```

```python
qiskit.quantum_info.Statevector(circuit()[1]).draw('bloch')
```
