```python
import pennylane as qml
from pennylane import numpy as np
import qiskit
import json
```

```python
n_qubits = 2
```

```python
dev = qml.device('default.qubit', wires=n_qubits)
```

```python
@qml.qnode(dev)
def circuit():
    qml.Hadamard(wires=0)
    qml.Hadamard(wires=1)
    qml.CZ(wires=[0, 1])
    qml.Hadamard(wires=1)
    return qml.state()
```

```python
qml.draw_mpl(circuit)()
```

```python
qiskit.quantum_info.Statevector(circuit()).draw('latex')
```
